```python


def e_gcd(a,b):
    assert(
        a>=0 and b>=0 )
    if b==0:
        return (a,1,0)
    (q,r) = divmod(a,b)
    (d,s,t) = e_gcd(b,r)
    # gcd(a, b) == gcd(b, r) == s*b + t*r == s*b + t*(a - q*b)
    return (d,t,s-q*t)


def test_e_gcd(n):
    for i in range(n):
        (d,s,t) = e_gcd(i,n)
        if d==1:
            print(i,"*",s,"= 1 mod",n)
            assert( i*s%n==1)
    for i in range(n):
        for j in range(n):
            print(i,j,e_gcd(i,j))
            pass

            
test_e_gcd(11)
```
